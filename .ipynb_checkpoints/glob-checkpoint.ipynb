```python
import numpy as np
# 全ての.pngファイルパスのリストを取得
all_png_list = glob('archive/gaussian_filtered_images/gaussian_filtered_images/*/*.png')
# 内包表記を使ってフォルダ名のリストを作成
type_list = [p.split('/')[3] for p in all_png_list]
# フォルダ名が取れているか確認
print(np.unique(type_list))
# 同様に拡張子のリストを作成
extension_list = [p.split('.')[-1] for p in all_png_list]
# 拡張子が取れているか確認
print(np.unique(extension_list))
```

```python
len(all_png_list)
```

```python
import pandas as pd
df = pd.DataFrame({'filepath': all_png_list, 'type': type_list, 'extension': extension_list})
df.head(10)
```

```python
df['type'].value_counts()
```

```python
import seaborn as sns
%matplotlib inline

sns.countplot(x='type',data=df)
```
