```python
x = np.linspace(-3,3,10)
y1 = np.exp(x)
y2 = np.exp(x)*2
plt.plot(x,y1, label='y1')
plt.plot(x,y2, label='y2')
plt.legend()
plt.axis('off')
```

```python
x = np.linspace(-3,3,10)
y1 = np.exp(x)
y2 = x*x


plt.subplot(1,2,1)
plt.plot(x,y1, label='y1')
plt.legend()

plt.subplot(1,2,2)
plt.plot(x,y2, label='y2')
plt.legend()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.plot(x,y1)
ax2.plot(x,y2)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 6))
axes[0,].plot(x, y1)
axes[1,].plot(x, x*x)
 
axes[0,].set_title('first')
axes[1,].set_title('second')
 
axes[0,].set_xlabel('x')
axes[0,].set_ylabel('y')
axes[1,].set_xlabel('x')
axes[1,].set_ylabel('y')
 
fig.savefig('savefig_sample.png')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 6))
axes[0,].plot(x, y1)
axes[1,].plot(x, x*x)
 
axes[0,].set_title('first')
axes[1,].set_title('second')
 
axes[0,].set_xlabel('x')
axes[0,].set_ylabel('y')
axes[1,].set_xlabel('x')
axes[1,].set_ylabel('y')

plt.tight_layout()

fig.savefig('savefig_sample2.png')
```

```python
fig, axes = plt.subplots()
axes.plot(x, y1, label='first')
axes.plot(x, y2, label='second')
axes.plot(x, y1+y2, label='first + second')
axes.legend()
```

```python
from matplotlib.backends.backend_pdf import PdfPages
pdf = PdfPages('savefig_sample.pdf')

fig, axes = plt.subplots()
axes.plot(x, y1, label='first')
axes.plot(x, y2, label='second')
axes.plot(x, y1+y2, label='first + second')
axes.legend()

pdf.savefig(fig)
pdf.close()
```

```python
pdf = PdfPages('savemultifig_sample.pdf')
for i in range(0, 10):
    #--------------
    fig, axes = plt.subplots()
    axes.plot(x, y1 + x*i)
    axes.set_title('ID:#{}'.format(i))
    pdf.savefig(fig)
 
pdf.close()
```
